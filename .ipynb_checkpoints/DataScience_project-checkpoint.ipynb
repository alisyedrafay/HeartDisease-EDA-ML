```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
data["Sex"].value_counts()
```

```python
data["ChestPainType"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
data
```

```python
data["ChestPainType"].value_counts()
```

```python
transformer = ColumnTransformer([
    ("tnf1",OneHotEncoder(sparse_output=False), ["Sex"] ),
    ("tnf1",OrdinalEncoder(categories=["ASY", "NAP", "ATA", "TA"]),["ChestPainType"])], remainder="passthrough")
```

```python

tranformer
```

```python
tranformer.named_transformers_["tnf2"]
```

```python
X_trained_trans = tranformer.fit_transform(X_train)
X_test_tran = tranformer.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
X_trained = scaler.fit_transform(X_trained_trans)
X_tested = scaler.transform(X_test_tran)
```

```python
np.round(X_train.describe(),1)
```

```python
X_trained = pd.DataFrame(X_trained, columns=data.columns)
X_tested = pd.DataFrame(X_tested, columns=data.columns)
```

```python
np.round(X_trained.describe(),1)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dt = DecisionTreeClassifier(criterion="entropy", max_depth=4)
dt.fit(X_trained, y_train)
```

```python
y_pred = dt.predict(X_tested)
y_pred
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score
```

```python
print("Accuarcy score is ",accuracy_score(y_pred, y_test)*100)
```

```python
plt.figure(figsize=(10,4))
plot_tree(dt, filled=True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf
```

```python
rf.fit(X_trained, y_train)
```

```python
y_prediction = rf.predict(X_tested)
y_prediction
```

```python
print("Accuarcy score is ", accuracy_score(y_prediction, y_test))
```
